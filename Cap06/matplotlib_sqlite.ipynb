# matplotlib

```python
import sqlite3
import matplotlib.pyplot as plt
%matplotlib notebook 
#comando que abre o grafico nessa janela
#https://matplotlib.org/tutorials/introductory/pyplot.html

def cria_grafico():
    cur.execute("SELECT id, valor FROM produto")
    ids = []
    valores = []
    dados = cur.fetchall()
    
    for linha in dados:
        ids.append(linha[0])
        valores.append(linha[1])

    plt.subplot(131)
    plt.bar(ids, valores)
    plt.subplot(132)
    plt.scatter(ids, valores)
    plt.subplot(133)
    plt.plot(ids, valores)
    plt.suptitle('Tipos de plotagem')
    plt.show()
```

```python
conn = sqlite3.connect('dsa.db')
cur = conn.cursor()
```

```python
cria_grafico()
```

```python
cur.close()
conn.close()
```
