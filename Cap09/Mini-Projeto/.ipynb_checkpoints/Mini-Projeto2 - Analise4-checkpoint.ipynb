## Análise 4

```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style="white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "dataset/autos/autos.csv"
df = pd.read_csv(clean_data_path,encoding="latin-1")
```

```python
# Calcule a média de preço por marca e por veículo
df.groupby(['brand', 'vehicleType'])['price'].mean().astype(int)
```

## Preço médio de um veículo por marca, bem como tipo de veículo

```python
# Crie um Heatmap com Preço médio de um veículo por marca, bem como tipo de veículo
df2 = df.copy()
mean = df2.groupby(['brand', 'vehicleType'])['price'].mean().astype(int)

# adicionando a coluna no dataframe
df2 = df.set_index(['brand', 'vehicleType'])
df2['avgPrice'] = mean
df2 = df2.reset_index()

# Escolhe os pivôs
htmap = pd.pivot_table(df2, index='brand', columns='vehicleType', values='avgPrice')

# Draw a heatmap with the numeric values in each cell
plt.figure(figsize=(15, 25))
sns.heatmap(htmap, annot=True, fmt=".0f", cmap="YlGnBu" )
plt.xlabel("Tipo de Veículo")
plt.ylabel("Marca")
plt.title("Preço médio de um veículo por marca, bem como tipo de veículo")
plt.show()
```
