```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style="white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "dataset/autos/autos.csv"
df = pd.read_csv(clean_data_path,encoding="latin-1")
```

## Número de veículos pertencentes a cada marca

```python
# Crie um Plot que mostre o número de veículos pertencentes a cada marca
sns.set_style("whitegrid")
sns.catplot(y="brand", data=df, kind="count", palette="Reds_r", height=7, aspect=1.5)
plt.xlabel("Número de Veículos")
plt.ylabel("Marca")
plt.title("Veículos por Marca")
plt.show()
```

## Preço médio dos veículos com base no tipo de veículo, bem como no tipo de caixa de câmbio

```python
# Crie um Plot com o Preço médio dos veículos com base no tipo de veículo, bem como no tipo de caixa de câmbio
sns.set(style="whitegrid")
sns.catplot(x="vehicleType", y="price", hue="gearbox", data=df, kind="bar", height = 7)
plt.xlabel("Tipo de Veículos")
plt.ylabel("Preço")
plt.title("Preço médio dos veículos por tipo de veículo e tipo de caixa de câmbio")
plt.show()
```
